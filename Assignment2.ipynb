```python
# CHECK AND REMOVE DUPLICATE RECORDS...
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Convert date columns to datetime...
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
# Convert numerical columns to int/float
```

```python
df['Annual Income'] = df['Annual Income'].astype(int)
df['Price ($)'] = df['Price ($)'].astype(float)
```

```python
#Fix casing issues (upper/lower/title case)...
```

```python
text_cols = ['Customer Name', 'Dealer_Name', 'Company', 'Model',
             'Engine', 'Transmission', 'Color', 'Body Style', 'Dealer_Region']

df[text_cols] = df[text_cols].apply(lambda x: x.str.strip().str.title())
```

```python
#Summary statistics using describe()
```

```python
df.describe()
```

```python
# Value counts for categorical columns...
```

```python
df['Body Style'].value_counts()
```

```python
# Group-by analysis (e.g., average, total, count)...
```

```python
df.groupby('Company')['Price ($)'].mean().sort_values(ascending=False)
```

```python
# Identify top or bottom performing categories
```

```python
df.groupby('Dealer_Region')['Price ($)'].sum().sort_values(ascending=False)
```

```python
# Correlation analysis between numerical columns
```

```python
df[['Annual Income', 'Price ($)']].corr()
```

```python
#Bar Chart – Category-wise comparison...
```

```python
df.groupby('Body Style')['Price ($)'].mean().plot(kind='bar', title='Average Price by Body Style')
plt.xlabel('Body Style')
plt.ylabel('Average Price')
plt.show()
```

```python
# Line Chart – Trend analysis
```

```python
df.groupby('Date')['Price ($)'].sum().plot(title='Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Histogram – Distribution of a numerical column
```

```python
plt.hist(df['Price ($)'], bins=20)
plt.title('Distribution of Car Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter Plot – Relationship between two variables
```

```python
plt.scatter(df['Annual Income'], df['Price ($)'])
plt.title('Annual Income vs Car Price')
plt.xlabel('Annual Income')
plt.ylabel('Car Price')
plt.show()
```

```python
# Heatmap – Correlation between numerical columns
```

```python
import seaborn as sns
sns.heatmap(df[['Annual Income', 'Price ($)']].corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
Insight1:-SUVs dominate the market in terms of sales volume, indicating strong consumer preference for larger and utility-focused vehicles.

Insight2:-Customers with higher annual income tend to purchase higher-priced vehicles, showing a clear relationship between purchasing power and car pricing.

Insight3:-Luxury brands such as Cadillac and Acura command significantly higher average prices, contributing more revenue per sale despite lower volume.

Insight4:-Sales trends over time show consistent demand with noticeable peaks, suggesting possible seasonal buying behavior.

Insight5:-Some dealer regions contribute disproportionately to total sales revenue, highlighting opportunities for region-focused marketing and inventory planning.
```
